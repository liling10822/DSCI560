# Random number generator

```python
import csv
import seaborn as sns
```

###  Invoking random_num.py for randomly generating 1000 numbers between 0 to 100 and resulting in random_num.csv file


```python
%run random_num.py
```

### Opening random_num.csv to generate a distrubution gragh of 1000 random numbers by using seaborn package

```python
with open('random_num.csv','r') as random_num:
    reader = csv.reader(random_num)
    random_num_list = list(reader)
fig = sns.distplot(random_num_list,color='red')
fig.set(xlabel='Range', ylabel='Possibility')
```

### invoking function_random_num.py for generating 1000 numbers based on the function y=3x+6 from random_num.csv, and resulting in fucntion_random_num.csv file


```python
%run function_random_num.py
```

### Opening function_random_num.csv to generate a distribution graph of 1000 numbers based on function y=3x+6 by using the seaborn package

```python
with open('function_random_num.csv','r') as function_random_num:
    reader = csv.reader(function_random_num)
    function_random_num_list = list(reader)
fig = sns.distplot(function_random_num_list,color='red')
fig.set(xlabel='Range', ylabel='Possibility')
```
